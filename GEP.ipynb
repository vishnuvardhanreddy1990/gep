```python
import pandas as ps
import matplotlib.pyplot as plt
```

```python
def read_file(path):
  """
  read file using pandas
  """
  file_data=ps.read_csv(path)
  file_data.head(5)
  file_data.replace('..', 0, inplace=True)
  return file_data
```

```python
df=read_file("/content/GEP.csv")
df_v=df.columns
df_v=df_v[4:]
```

```python
df
```

```python
def line(df_v):
  """
  here we create line plot
  """
  df[df_v] = df[df_v].astype(float)
  df.plot(x='Series Code', y=df_v, kind='line', color='red')
  plt.xlabel('Series Code')
  plt.ylabel(df_v)
  plt.title('Value by Series Code')
  plt.show()
```

```python
def hist_plot(df_v):
  """
  here we create hist plot
  """
  df[df_v] = df[df_v].astype(float)
  df.plot(x='Series Code', y=df_v, kind='hist', color='red')
  plt.xlabel('Series Code')
  plt.ylabel(df_v)
  plt.title('Value by Series Code')
  plt.show()
```

```python
def fun_plot(df_v):
  """
  here we create plot
  """
  df[df_v] = df[df_v].astype(float)
  df.plot(x='Series Code', y=df_v, color='red')
  plt.xlabel('Series Code')
  plt.ylabel(df_v)
  plt.title('Value by Series Code')
  plt.show()
```

```python
def box_plot(df_v):
  """
  here we create box plot
  """
  df[df_v] = df[df_v].astype(float)
  df.plot(x='Series Code', y=df_v, kind='box', color='red')
  plt.xlabel('Series Code')
  plt.ylabel(df_v)
  plt.title('Value by Series Code')
  plt.show()
```

```python
def bar_plot(df_v):
  """
  here we create bar plot
  """
  df[df_v] = df[df_v].astype(float)
  df.plot(x='Series Code', y=df_v, kind='bar', color='red')
  plt.xlabel('Series Code')
  plt.ylabel(df_v)
  plt.title('Value by Series Code')
  plt.show()
```

```python
for i in df_v:
  """
  here we call columns one by one
  """
  print(i)
  bar_plot(i)
  box_plot(i)
  fun_plot(i)
  line(i)
  hist_plot(i)
```
